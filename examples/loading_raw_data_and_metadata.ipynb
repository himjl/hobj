## Loading image meta

The cells below show you how to load the image metadata for the two experiments. This is not necessary for scoring a model, but you might have another reason for doing this.

```python
df_meta_highvar = hobj.load_imageset_meta_highvar()
df_meta_highvar
```

```python
df_meta_oneshot = hobj.load_imageset_meta_oneshot()
df_meta_oneshot
```

```python
df_meta_catch = hobj.load_imageset_meta_catch() # These are used for catch trials in Experiment 1.
df_meta_warmup = hobj.load_imageset_meta_warmup() # These were used to 'warm up' the human participants, and also to normalize image features prior to benchmarking.
```

## Loading images

`hobj` offers the `load_image` function for loading images by `image_id` as `PIL.Images`, and the `get_image_path` function as an alternative. See below!

```python
image_id = df_meta_highvar['image_id'].iloc[3]
image = hobj.load_image(image_id = image_id)
image
```

```python
hobj.get_image_path(image_id=image_id)
```
